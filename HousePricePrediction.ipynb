```python
# " House Price Prediction using Linear Regression "
# This notebook predicts house prices based on area using Python and scikit-learn.
```

```python
# Imported libraries

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# We load the dataset containing house area and price

data=pd.read_csv('data-houseprice.csv')
data.head()
```

```python
# We train a linear regression model using area as the predictor variable

# Defined features and target
x=data[['area']]
y=data[['price']]

# Train linear regression model
model=LinearRegression()
model.fit(x,y)

# Predictions
pre_price=model.predict(x)
pre_price
```

```python
# The scatter plot shows actual data points, and the red line represents the regression fit

plt.scatter(x,y,color='skyblue')
plt.plot(x, pre_price, color='red')
plt.title("House Price Prediction")
plt.xlabel("area (sq. feet)")
plt.ylabel("price (rupees)")
plt.show()
```

```python
# We test the model by predicting the price of a house with sq. feet area

n = int(input("Enter the area (in sq feet): "))
predicted_price = model.predict(pd.DataFrame([[n]], columns=['area']))[0][0]
(f"{predicted_price:.2f} is the price of the house...")
```

```python
# The R2 score indicates how well the model fits the data

("Model R2 score:", r2_score(y, pre_price))
```
